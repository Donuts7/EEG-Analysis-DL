```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('DREAMR.csv')
```

```python
dataset.drop(columns=['Unnamed: 0'],inplace=True)
dataset = dataset.T
```

```python
dataset.head()
```

```python
print(dataset.shape)
```

```python
labels = dataset.drop(columns=list(range(0,107520)))
data = dataset.drop(columns=list(range(107520,107523)))
```

```python
#zi = (xi – min(x)) / (max(x) – min(x)) * 100

min_val = min(data.min())
max_val = max(data.max())

data = (data - min_val) / (max_val - min_val) * 255
data = round(data)
```

```python
data
```

```python
from PIL import Image
for i in range(0,414):
    img = data.iloc[i].to_numpy(dtype=np.uint8)
    img = img.reshape(280,384)
    image = Image.fromarray(img)
    name = str(i)+'img.png'
    image.save('img_dataset/'+name)

#import matplotlib.pyplot as plt
#plt.imshow(image,'gray')
```
